```python
def getAverageAge(Pclass):
    return round(titanic_df[titanic_df['Pclass'] == Pclass]['Age'].mean(), 0)
```

```python
for i in titanic_df.index:
    if pd.isnull(titanic_df['Age'][i]):
        if titanic_df['Pclass'][i] == 1:
            titanic_df['Age'][i] = getAverageAge(1)
        elif titanic_df['Pclass'][i] == 2:
            titanic_df['Age'][i] = getAverageAge(2)
        elif titanic_df['Pclass'][i] == 3:
            titanic_df['Age'][i] = getAverageAge(3)
    else: 
        continue
```
